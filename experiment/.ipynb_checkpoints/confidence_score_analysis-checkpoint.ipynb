```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
```

```python
df = pd.read_json("data.json")
df.drop_duplicates(subset='analyzed_material', keep='last', inplace=True)
df.reset_index(drop=True, inplace=True)
print(len(df['analyzed_material']))
df.head()

newdf = pd.read_json("data5.json")
```

```python
color_mapping = {"PFAS (No)": "green", "PFAS (Yes)": "red", "PFAS (Undetermined)": "orange", "Undetermined": "yellow"}
colors = df["decision"].map(color_mapping)
plt.scatter(df.index, df["confidence"], c=colors)
plt.figure(figsize=(20,12))
plt.show()
```

```python
color_mapping = {"PFAS (No)": "green", "PFAS (Yes)": "red", "PFAS (Undetermined)": "orange", "Undetermined": "yellow"}
colors = newdf["decision"].map(color_mapping)
plt.scatter(newdf.index, newdf["confidence"], c=colors)
plt.figure(figsize=(20,12))
plt.show()
```

```python
custom_color_sequence = ["green", "blue", "red"]
fig = px.scatter(
    newdf, 
    x = newdf.index,
    y = "confidence",
    color="decision",
    title="Confidence Scores in experiment run",
    labels={"decision": "decision", "confidence": "confidence", "material_name": "analyzed_material"},
    color_discrete_sequence=custom_color_sequence
)
fig.update_layout(width=1400, height=600, plot_bgcolor="white", paper_bgcolor="black", showlegend=True)
fig.update_xaxes(showgrid=False)
fig.update_yaxes(showgrid=True, gridwidth=0.5, gridcolor='grey')
mean_conf = np.mean(newdf["confidence"])
fig.add_trace(
    go.Scatter(
        x=[newdf.index.min(), newdf.index.max()],
        y=[mean_conf, mean_conf],
        mode="lines",
        line=dict(color="purple", width=2),
        name=f'Mean Confidence: {mean_conf:.2f}'
    )
)
```

```python
fig
```

```python
# print(f"Median: {np.median(df["confidence"])}")
# print(f"Standard Deviation: {np.std(df["confidence"])}")
q1 = np.percentile(df["confidence"], 25)
q3 = np.percentile(df["confidence"], 75)
iqr = q3 - q1
print(f"IQR: {q3-q1}")
MAD = np.mean(np.abs(df['confidence'] - np.mean(df['confidence'])))
print(f"MAD: {MAD}")
outliers = np.mean(df["confidence"]) - 1.5*iqr
print(outliers)
print(np.mean(df["confidence"]))
```

```python
# print(f"Median: {np.median(df["confidence"])}")
# print(f"Standard Deviation: {np.std(df["confidence"])}")
q1 = np.percentile(newdf["confidence"], 25)
q3 = np.percentile(newdf["confidence"], 75)
iqr = q3 - q1
print(f"IQR: {q3-q1}")
MAD = np.mean(np.abs(newdf['confidence'] - np.mean(newdf['confidence'])))
print(f"MAD: {MAD}")
outliers = np.mean(newdf["confidence"]) - 1.5*iqr
print(outliers)
print(np.mean(newdf["confidence"]))
```

```python
fig.add_trace(
    go.Scatter(
        x=[df.index.min(), df.index.max()],
        y=[outliers, outliers],
        mode="lines",
        line=dict(color="slateblue", width=2, dash="dash"),
        name=f'Outliers line: {outliers:.5f}'
    )
)
fig
```
